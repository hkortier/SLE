```python
%matplotlib inline

from ipywidgets import widgets
#from IPython.html.widtgets import *


import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0.0,10.0,200)
```

```python
def plotSine(frequency,amplitude):
    f = frequency
    a = amplitude 
    plt.figure(figsize=(12,8))
    x = a*np.sin(2*np.pi*f*t)
    plt.plot(t,x)
    
    plt.xlabel('Tijd (t)')
    plt.ylabel('Amplitude (V)')
    plt.grid()
    plt.axis([0,10,-1,1])
    #plt.show()
```

```python
def plotFilteredSine(f,C):
    plt.figure(figsize=(12,8))
    x = np.sin(2*np.pi*f*t)
    plt.plot(t,x)
    xFilt = lowPassFilteredSignal(f,C)
    plt.plot(t,xFilt)
    
    plt.xlabel('Tijd (t)')
    plt.ylabel('Amplitude (V)')
    plt.legend(('Ingang','Uitgang'))
    plt.grid()
    plt.axis([0,10,-1,1])
```

```python
def lowPassFilteredSignal(f,C=1):
    R = 2e-1
    modulus = 1/np.sqrt(1+2*np.pi*f*R*C)
    phaseShift = -np.arctan(2*np.pi*f*R*C)
    return modulus*np.sin((2*np.pi*f*t+phaseShift))
```

```python
widgets.interact(plotSine, frequency=(0.1,1,0.1), amplitude=(0.1,1,0.1))
```

```python
widgets.interact(plotFilteredSine, f=(0.1,1,0.1), C = (1e-1,1e1,1))
```
